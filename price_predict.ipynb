```python
x = car.drop(columns="selling_price")
y = car["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[["name","year","km_driven","fuel"]])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_),["name","year","km_driven","fuel"]),remainder = "passthrough")
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```

```python
scores = []
for i in range(1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
import numpy as np
np.argmax(scores)
```

```python
scores[531]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open("final_model.pkl","wb"))
```

```python
pipe.predict(pd.DataFrame([["Hyundai Verna 1.6 SX",2012,10000,"Diesel",]],columns = x_test.columns))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
pipe2 = make_pipeline(column_trans,rf)
pipe2.fit(x_train,y_train)
y_pred2 = pipe2.predict(x_test)
r2_score(y_test,y_pred2)
```
